```python
import pandas as pd
from pathlib import Path

# Chargement robuste
data_path = Path("data") / "prices_round_5_day_2.csv"
df = pd.read_csv(data_path, sep=";")

# Reste inchangé
df['mid_price'] = (df['bid_price_1'] + df['ask_price_1']) / 2
group_mids = df.groupby(['timestamp', 'product'])['mid_price'].mean().unstack()

for lag in [1, 3, 5]:
    print(f"\n--- Corrélation avec Lag de {lag} ticks ---")
    shifted_data = group_mids.shift(lag)
    corr_matrix = group_mids.corrwith(shifted_data, axis=0)
    print(corr_matrix.sort_values(ascending=False).head(10))
```

```python
import pandas as pd
from pathlib import Path

# Chargement robuste
data_path = Path("data") / "prices_round_5_day_4.csv"
df = pd.read_csv(data_path, sep=";")

# Reste inchangé
df['mid_price'] = (df['bid_price_1'] + df['ask_price_1']) / 2
group_mids = df.groupby(['timestamp', 'product'])['mid_price'].mean().unstack()

for lag in [1, 3, 5]:
    print(f"\n--- Corrélation avec Lag de {lag} ticks ---")
    shifted_data = group_mids.shift(lag)
    corr_matrix = group_mids.corrwith(shifted_data, axis=0)
    print(corr_matrix.sort_values(ascending=False).head(10))
```

```python
# On utilise tes données déjà chargées dans group_mids
# On calcule la moyenne du groupe CHIPS et du groupe ROBOTS
chips_trend = group_mids[["MICROCHIP_CIRCLE", "MICROCHIP_OVAL", "MICROCHIP_SQUARE", "MICROCHIP_RECTANGLE", "MICROCHIP_TRIANGLE"]].mean(axis=1)
robots_trend = group_mids[["ROBOT_VACUUMING", "ROBOT_MOPPING", "ROBOT_DISHES", "ROBOT_LAUNDRY", "ROBOT_IRONING"]].mean(axis=1)

# TEST 1 : Est-ce que CHIPS mène ROBOTS ? 
# On décale CHIPS vers le futur pour voir s'il prédit ROBOTS
corr1 = robots_trend.corr(chips_trend.shift(1)) 

# TEST 2 : Est-ce que ROBOTS mène CHIPS ?
corr2 = chips_trend.corr(robots_trend.shift(1))

print(f"Score CHIPS mène ROBOTS : {corr1}")
print(f"Score ROBOTS mène CHIPS : {corr2}")

if corr1 > corr2:
    print("✅ VERDICT : CHIPS est le Leader. On achète les ROBOTS quand les CHIPS montent.")
```

```python
GROUPS = {
    "GALAXY": ["GALAXY_SOUNDS_DARK_MATTER", "GALAXY_SOUNDS_BLACK_HOLES", "GALAXY_SOUNDS_PLANETARY_RINGS", "GALAXY_SOUNDS_SOLAR_WINDS", "GALAXY_SOUNDS_SOLAR_FLAMES"],
    "SLEEP": ["SLEEP_POD_SUEDE", "SLEEP_POD_LAMB_WOOL", "SLEEP_POD_POLYESTER", "SLEEP_POD_NYLON", "SLEEP_POD_COTTON"],
    "CHIPS": ["MICROCHIP_CIRCLE", "MICROCHIP_OVAL", "MICROCHIP_SQUARE", "MICROCHIP_RECTANGLE", "MICROCHIP_TRIANGLE"],
    "PEBBLES": ["PEBBLES_XS", "PEBBLES_S", "PEBBLES_M", "PEBBLES_L", "PEBBLES_XL"],
    "ROBOTS": ["ROBOT_VACUUMING", "ROBOT_MOPPING", "ROBOT_DISHES", "ROBOT_LAUNDRY", "ROBOT_IRONING"],
    "VISORS": ["UV_VISOR_YELLOW", "UV_VISOR_AMBER", "UV_VISOR_ORANGE", "UV_VISOR_RED", "UV_VISOR_MAGENTA"],
    "TRANSLATORS": ["TRANSLATOR_SPACE_GRAY", "TRANSLATOR_ASTRO_BLACK", "TRANSLATOR_ECLIPSE_CHARCOAL", "TRANSLATOR_GRAPHITE_MIST", "TRANSLATOR_VOID_BLUE"],
    "PANELS": ["PANEL_1X2", "PANEL_2X2", "PANEL_1X4", "PANEL_2X4", "PANEL_4X4"],
    "SHAKES": ["OXYGEN_SHAKE_MORNING_BREATH", "OXYGEN_SHAKE_EVENING_BREATH", "OXYGEN_SHAKE_MINT", "OXYGEN_SHAKE_CHOCOLATE", "OXYGEN_SHAKE_GARLIC"],
    "SNACKS": ["SNACKPACK_CHOCOLATE", "SNACKPACK_VANILLA", "SNACKPACK_PISTACHIO", "SNACKPACK_STRAWBERRY", "SNACKPACK_RASPBERRY"]
}
```

```python
# 1. Calculer les moyennes de prix pour TOUS les groupes
group_data = {}
for g_name, g_prods in GROUPS.items():
    group_data[g_name] = group_mids[g_prods].mean(axis=1)

groups_df = pd.DataFrame(group_data)

# 2. Tester toutes les combinaisons avec un décalage de 1 tick
print("--- Recherche du couple Leader-Suiveur (Lag 1) ---")
lags = [1, 2, 3]
results = []

for lag in lags:
    for leader in groups_df.columns:
        for follower in groups_df.columns:
            if leader == follower: continue
            corr = groups_df[follower].corr(groups_df[leader].shift(lag))
            results.append({'Leader': leader, 'Follower': follower, 'Lag': lag, 'Corr': corr})

# Afficher les 10 meilleurs signaux trouvés
results_df = pd.DataFrame(results)
print(results_df.sort_values(by='Corr', ascending=False).head(10))
```

```python
# Tester par exemple au sein du groupe PEBBLES
group_name = "PEBBLES"
prods = GROUPS[group_name]

print(f"--- Recherche du Leader au sein de {group_name} ---")
for p1 in prods:
    for p2 in prods:
        if p1 == p2: continue
        corr = group_mids[p2].corr(group_mids[p1].shift(1))
        if corr > 0.90: # On cherche une corrélation très forte
            print(f"{p1} mène {p2} avec une corrélation de {corr:.4f}")
```

```python
# On calcule les variations (diff)
trans_diff = groups_df["TRANSLATORS"].diff().dropna()
visors_diff = groups_df["VISORS"].diff().dropna()

# Calcul du Beta (pente de la relation)
# Beta = Covariance(X,Y) / Variance(X)
beta = visors_diff.cov(trans_diff) / trans_diff.var()

print(f"--- Résultat du Beta ---")
print(f"Pour 1 point de mouvement chez TRANSLATORS, VISORS bouge de : {beta:.4f} points")
```

```python
# On choisit un produit suiveur (ex: UV_VISOR_YELLOW)
product_name = "UV_VISOR_YELLOW"
group_mean = group_mids[GROUPS["VISORS"]].mean(axis=1)

# Calcul de l'écart (Spread)
spread = group_mids[product_name] - group_mean

print(f"--- Analyse du Spread pour {product_name} ---")
print(f"Écart moyen : {spread.mean():.4f}")
print(f"Écart type (Std Dev) : {spread.std():.4f}")

# Si l'écart type est faible (ex: 2.0), tu peux acheter quand spread < -4 et vendre quand spread > 4
```
